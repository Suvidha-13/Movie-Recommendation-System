```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('final_data.csv')
```

```python
df
```

```python
df['combined'] = df['combined'].apply(lambda x:x.lower())
```

```python
df['combined'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
df['combined'] = df['combined'].apply(stem)
```

```python
df['combined'][0]
```

```python
df['combined'][1]
```

```python
%pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
df.shape
```

```python
vectors = cv.fit_transform(df['combined']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = df[df['movie_title'] == movie.lower()].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(df.iloc[i[0]].movie_title)
```

```python
recommend("iron man")
```
